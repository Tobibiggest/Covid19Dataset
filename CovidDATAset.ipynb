```python
# splitting the cvariables into categorical and numerical variable

cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
# checking for skewness and outliers in our numeric variable 

for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

# OBSERVATIONS;

 * Deaths - right skewed on 6.94.
 * Recovered - right skewed on 6.93.

# DATA TRANSFORMATION

 * Our Numerical variables are highly skewed, hence we use logh transformation to Normalize them.

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])

            
data.info()
```

```python
log_transform(data,['Recovered','Deaths', 'Confirmed'])
```
